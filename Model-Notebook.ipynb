```python
import math 
def plot_all_hist(df):
    numeric_cols = df.select_dtypes(exclude=["O"]).columns
    n_cols = 3
    n_rows = math.ceil(len(numeric_cols) / n_cols)

    plt.figure(figsize=(5*n_cols,4*n_rows))

    for i, col in enumerate(numeric_cols,1):
        plt.subplot(n_rows,n_cols,i)
        sns.histplot(df[col],kde=True,bins=50)

    plt.tight_layout()
    plt.show()
    
```

```python
plot_all_hist(df)
```

```python
import math
def all_box_plot(df):
    categorical_columns = df.select_dtypes(include = ["O"]).columns
    n_cols = 3
    n_rows = math.ceil(len(categorical_columns) / n_cols)

    plt.figure(figsize=(5*n_cols, 5*n_rows))

    for i, cols in enumerate(categorical_columns,1):
        plt.subplot(n_cols,n_rows,i)
        sns.boxplot(data=df,x=cols,y="Price")

    plt.tight_layout()
    plt.show()
```

```python
all_box_plot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Price",axis=1)
y = df["Price"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
X_train.head()
```

```python
from deneme import CarPriceEncoder
```

```python
encoder = CarPriceEncoder()
```

```python
X_train_encoded = encoder.fit_transform(X_train,y_train)
```

```python
X_test_encoded = encoder.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train_encoded,y_train)
```

```python
y_pred = model.predict(X_test_encoded)
```

```python
print("r2:", r2_score(y_test,y_pred))
```

```python
rf_params = {
    "n_estimators": [100,200,500,1000],
    "max_depth": [5,8,10,15,None],
    "max_features": ["sqrt","log2",5,7,10],
    "min_samples_split": [2,8,12,20],
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rscv = RandomizedSearchCV(estimator=RandomForestRegressor(),param_distributions=rf_params,n_jobs=-1,n_iter=100,cv=5)
```

```python
rscv.fit(X_train_encoded,y_train)
```

```python
rscv.best_params_
```

```python
model = RandomForestRegressor(n_estimators=500,min_samples_split=2,max_features=10,max_depth=None)
model.fit(X_train_encoded,y_train)
y_pred = model.predict(X_test_encoded)
print("r2:", r2_score(y_test,y_pred))
```
